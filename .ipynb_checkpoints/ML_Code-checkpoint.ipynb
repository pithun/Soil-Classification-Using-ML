## NOISE IN IMAGES

There's some common types of noise namely;
1. Salt and Pepper - Adding dark and white pixels to positions in the image
2. Gaussian noise - randomly generated pixels from a gaussian distribution to the picture. In the test I did, the images are already uint8 so, it was okay for me to say 0 is black and 255 is the whitest pixel and **plt.imshow shows that if we wanted to set our own range, imshow has some parameters like vmin and vmax which we can use actually, imshow takes the lowest as black and highest as white and uses a linear way to distribute colors**. 
- But in Gaussian noise, normally distributed noise (with mean 0) we'd have some -ve pixels and some +ve pixels we can't say 0 is the darkest instead 0 is the mean which is inbetween and can be said to be gray. Basically, we define our own darkest pixel say -20 and whitest say 20 now after genrating random numbers between 0 and 1 following a standard normal distribution if we multiply by some sigma which in this case, optimal will be 20, the whiteness of individual pixel values will increase.
- The ideas basially is that we follow a standard normal distribution to generate values and we set our own min and max and then as we multiply the generated number by values close to max, we notice an increase in whiteness.

Test

2. Add gausian noise to the image

```python
test_img_shape = test_image_gray.shape
rows=test_img_shape[0]
columns=test_img_shape[1]
```

```python
# The idea is to make some positions 1, some 0, some 255
salt_pepper = np.ones(test_img_shape)

# First we loop through every row
for row in range(rows):
    # We use randint to generate 800 random numbers indicating 800 random positions for salt and 600 for pepper all together
    # 1400 or less positions on a single row. Less because the second rand list might take positions of the first
    random_salt=np.random.randint(0,3024,800)
    random_pepper=np.random.randint(0,3000,600)
    salt_pepper[row,random_salt]=255
    salt_pepper[row,random_pepper]=0
```

```python
# Displaying to be sure it works
salt_pepper[0, :200]
```

```python
# Superimposing the gray image pixels to make some pixels black and some white
salt_and_pepper = test_image_gray * salt_pepper

# Some thresholding since other pixels are left the way they are. To retain similar meaning on a linear level, the max has to
# be the same as it was in the beginning other wise images that have been increased by the multiplication will reach value of
# max 255*255. Hence, values like 200 will be dark pixels
salt_and_pepper[salt_and_pepper[:,:]>=255]=255
```

```python
# Displays to be sure it works properly
test_image_gray[0, :30]
```

```python
salt_and_pepper[0,:30]
```

```python
# Displaying salt and pepper only
plt.imshow(salt_pepper, cmap='gray')
```

```python
# Displaying salt and pepper image
plt.imshow(salt_and_pepper, cmap='gray')
#cv.imwrite('my_salt_and_pepper.jpg', salt_and_pepper)
```

## Blurring Images

I've actually been trying to understand how blurring happens and I think what happens is that since convolution is a moving average basically in some large local neighbourhood you're averaging close pixels so it's kinda like you're spreading a single pixel intensity to a larger area. Say we have 5 pixel values **[0 0 9 0 0]** and 9 is like whitest pixel and 0 is darkest we can see that this is just an image with 1 white dot now say we take average with a 1x3 row kernel and 0 padding we'd have **[0 3 3 3 0]**. We'll see that we've made a more or less stretched gray image and can't see a pointed white. 
- **This is blurring-> stretching an image to loose pointed infomation**
